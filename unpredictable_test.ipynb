# Test of Hevensh's New Loss Function

test on the data in 482 points. compare with mse.

```python
!git clone https://github.com/Hevensh/HvsPredictor.git
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('HvsPredictor/price_data.csv')
data = np.array(data)
data = np.log(data)
data.resize([482,])
```

```python
from HvsPredictor import MyPredictor as hvs
import IPython.display

md1 = hvs.MyPredictor(data,metrics ='profit')
md1.deTrend()
md1.Options_set()
md1.compile_and_fit()
IPython.display.clear_output()
```

```python
md2 = hvs.MyPredictor(data,metrics='profit')
md2.deTrend()
md2.Options_set(loss_func='mse')
md2.compile_and_fit()
IPython.display.clear_output()
```

```python
md1.result_plot()
```

```python
md2.result_plot()
```
